# Star Wars Survey Clean
This dataset requires some cleaning as it appears to have used a checkbox system for participants to answer each question, as well as a ranking system to rank both movies and characters, making for a dataset that is quite messy.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

star_wars = pd.read_csv('Datasets/StarWars.csv', encoding = 'ISO-8859-1')
print(star_wars.columns)
```

```python
star_wars.head(10)
```

```python
star_wars.iloc[:,9:15].head(5)
```

## Convert Necessary Columns to Boolean

```python
bool_map = {
    'Yes': True,
    'No': False
}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(bool_map)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(bool_map)
star_wars.head(10)
```

```python
star_wars.iloc[1]
```

```python
star_wars.iloc[:, 3:9]
```

```python
movie_seen_map = {
    'Star Wars: Episode I  The Phantom Menace': True,
    'Star Wars: Episode II  Attack of the Clones': True,
    'Star Wars: Episode III  Revenge of the Sith': True,
    'Star Wars: Episode IV  A New Hope': True,
    'Star Wars: Episode V The Empire Strikes Back': True,
    'Star Wars: Episode VI Return of the Jedi': True,
    np.NaN: False
}

for i in range(3,9):
    star_wars.iloc[:, i] = star_wars.iloc[:, i].map(movie_seen_map)

star_wars.iloc[:, 3:9]
```

```python
star_wars = star_wars.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    'Unnamed: 4': 'seen_2',
    'Unnamed: 5': 'seen_3',
    'Unnamed: 6': 'seen_4',
    'Unnamed: 7': 'seen_5',
    'Unnamed: 8': 'seen_6'
})
star_wars.iloc[:, 3:9]
```

## Convert Necessary Columns to Float

```python
star_wars.iloc[:, 9:15]
```

```python
star_wars.iloc[1:, 9:15] = star_wars.iloc[1:, 9:15].astype(float)
star_wars.iloc[1:, 9:15]
```

```python
star_wars = star_wars.rename(columns={
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': "ranking_1",
    'Unnamed: 10': 'ranking_2',
    'Unnamed: 11': 'ranking_3',
    'Unnamed: 12': 'ranking_4',
    'Unnamed: 13': 'ranking_5',
    'Unnamed: 14': 'ranking_6'
})

star_wars.iloc[:, 9:15]
```

```python
star_wars.iloc[0]
```

## Finding the Highest-Ranked Movie

```python
mean_rankings = star_wars.iloc[1:, 9:15].mean()
mean_rankings.plot.bar()
```

The value of the ranking indicates how much people liked a particular film on average, with a low score indicating higher favourability. Here, we can see that the last three episodes are preferred more to the first three episodes.
## Finding the Most Viewed Movie

```python
movie_views = [star_wars[f'seen_{i}'].sum() for i in range(1,7)]
movie_views
```

## Cleaning Character Columns

```python
star_wars.iloc[:,15:29]
```

```python
star_wars['Unnamed: 16'].unique()
```

```python
star_wars= star_wars.rename(columns = star_wars.iloc[0,15:29])
star_wars.iloc[:,15:29]
```

```python
character_favour_map = {
    'Very favorably': 1,
    'Somewhat favorably': 2,
    'Neither favorably nor unfavorably (neutral)': 3,
    'Somewhat unfavorably': 4,
    'Very unfavorably': 5,
    'Unfamiliar (N/A)': np.NaN
}

for i in range(15,29):
    star_wars.iloc[:, i] = star_wars.iloc[:, i].map(character_favour_map)

star_wars.iloc[:, 15:29]
```

By converting the favourability rankings into numerical values, this allows for assessment of the average ranking for each known character. By converting the unfamiliar column to an NaN value, this allows for calculations to only consider favourability when taking into account whether the character is actually known.

```python
character_mean_rankings = star_wars.iloc[1:, 15:29].mean()
character_mean_rankings.plot.bar()
plt.xticks(rotation = 70)
plt.show()
```

The following bar plot reveals that Yoda, Han Solo and Obi Wan Kenobi are the most beloved characters from the Star Wars franchise amongst survey participants. In contrast, Jar Jar Binks, Emperor Palpatine and Boba Fett are among the least favoured characters.

```python
character_age_star_wars = star_wars.groupby('Age').agg(np.mean)
character_age_star_wars = character_age_star_wars.iloc[:, 7:]
character_age_star_wars
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize = (24, 13))
axes = [ax1, ax2, ax3, ax4]
colours = ['red', 'blue', 'green', 'purple']
for i in range(4):
    axes[i].barh(character_age_star_wars.columns, character_age_star_wars.iloc[i], color = colours[i])
    axes[i].set_xticks([1,2,3,4])
    axes[i].text(s = f'Age Bracket: {character_age_star_wars.index[i]}', x = 1.45, y = 14.5, weight = 'bold', size = 15)
    
```

After splitting character favourability by age bracket, we see little distinction between the character favourites, other than Luke Skywalker entering the top 3 most favoured characters in survery respondents 45+. Respondents under 45 have Yoda present in their top 3 instead. Respondents 45+ have the same bottom 3 characters: Jar Jar Binks, Boba Fett, and Emperor Palpatine. Respondents under 45 have Padme Amidala present in their bottom 3 instead of Boba Fett. This disparity between the two subgroups may be of significance considering that the first Star Wars movie was released in 1977, 46 years ago. Respondents 45+ are likely to have grown up watching all 6 movies throughout their youth and into adulthood. In contrast, respondents under 45 would have grown up during the release of Episodes I to III, potentially altering their perception of the characters. Analysis into the correlation between character favourability and their franchise introduction relative to respondent age may provide interesting results.
